```python

from IPython import get_ipython
from IPython.display import display
# %%
!pip install patoolib==1.1.1
from tensorflow.keras import layers, models
from google.colab import drive
import patoolib
from zipfile import ZipFile
import tensorflow as tf

# importing image files from dataset:
import os
patoolib.extract_archive("/content/drive/MyDrive/Hood College/archive.zip", outdir="/content/extracted_images")

# Specify the directory containing the extracted images
image_dir = '/content/extracted_images'

# Load and preprocess the images
image_dataset = tf.keras.utils.image_dataset_from_directory(
    image_dir,
    labels='inferred',  # Infer labels from subdirectory names
    label_mode='categorical',  # Use categorical labels
    image_size=(227, 227),  # Resize images to match AlexNet input shape
    interpolation='nearest',
    batch_size=16,  # Adjust batch size as needed
    shuffle=True  # Shuffle the dataset
)
# Preprocess the images (e.g., normalization)
def preprocess(image, label):
    image = tf.image.convert_image_dtype(image, dtype=tf.float32)  # Convert to float32
    image = tf.image.per_image_standardization(image)  # Normalize
    # Ensure the label is of type int32
    label = tf.cast(label, dtype=tf.int32)
    return image, label

image_dataset = image_dataset.map(preprocess)

# Split the dataset into training and validation sets
train_size = int(0.8 * len(image_dataset))
train_dataset = image_dataset.take(train_size)
validation_dataset = image_dataset.skip(train_size)












# Define the AlexNet architecture
def build_alexnet(input_shape=(227, 227, 3), num_classes=2):  # Change num_classes to 2
  model = models.Sequential([
  layers.Conv2D(96, (11, 11), strides=4, activation='relu', input_shape=input_shape),
  layers.MaxPooling2D((3, 3), strides=2),
  layers.Conv2D(256, (5, 5), padding='same', activation='relu'),
  layers.MaxPooling2D((3, 3), strides=2),
  layers.Conv2D(384, (3, 3), padding='same', activation='relu'),
  layers.Conv2D(384, (3, 3), padding='same', activation='relu'),
  layers.Conv2D(256, (3, 3), padding='same', activation='relu'),
  layers.MaxPooling2D((3, 3), strides=2),
  layers.Flatten(),
  layers.Dense(4096, activation='relu'),
  layers.Dropout(0.5),
  layers.Dense(4096, activation='relu'),
  layers.Dropout(0.5),
  layers.Dense(num_classes, activation='softmax') # Output layer for number of classes
])
  return model
# Build and compile the AlexNet model
alexnet_model = build_alexnet()
alexnet_model.compile(optimizer='adam', loss='categorical_crossentropy',
metrics=['accuracy'])
# Summary of the AlexNet model
alexnet_model.summary()

# Train the AlexNet model
alexnet_history = alexnet_model.fit(train_dataset, epochs=10, validation_data=validation_dataset)
```
