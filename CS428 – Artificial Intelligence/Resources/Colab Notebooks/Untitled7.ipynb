```python


import matplotlib.pyplot as plt

# Data
n_values = [10, 20, 30, 40]
execution_times = [0, 0, 3, 372]  # Execution times in milliseconds

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(n_values, execution_times, marker='o', linestyle='-', color='blue', label='Execution Time')

# Add labels and title
plt.xlabel('Input Size n')
plt.ylabel('Execution Time (ms)')
plt.title('Execution Time vs Input Size n for Recursive Fibonacci Function')
plt.legend()

# Optional: Add a grid
plt.grid(True)

# Show the plot
plt.show()
```
